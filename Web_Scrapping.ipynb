## __Beautiful Soup__

```
import bs4 as bs 
import urllib.request
import pandas as pd
import csv
```

```
sauce = urllib.request.urlopen('https://finviz.com/news.ashx').read()
```

```
soup = bs.BeautifulSoup(sauce, 'lxml')
```

```
print(soup.find_all('a'))
```

```
for paragraph in soup.find_all('a'): 
  print(paragraph.text)
```

```
for url in soup.find_all('a'):
  print(url.get('href'))
```

```
body = soup.body
```

```
for div in soup.find_all('div', class_='container'):
  print(div.text)
```

```
table = soup.table
print(table)
```

```
table_rows = table.find_all('tr')

for tr in table_rows:
  td = tr.find_all('td')
  row = [i.text for i in td]
  print(row)
```

# __Use Selenium for Web Crawling__

```
# install chromium, its driver, and selenium

!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
!pip install selenium
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```
# open it, go to a website, and get results
wd = webdriver.Chrome('chromedriver',options=options)
url = wd.get("http://econpy.pythonanywhere.com/ex/001.html")
```

```
print(wd.page_source)  # results
```

```
# Extract lists of "buyers" and "prices" based on xpath.
buyers = wd.find_elements_by_xpath(xpath='//div[@title="buyer-name"]')
prices = wd.find_elements_by_xpath('//span[@class="item-price"]')

# Print out all of the buyers and prices on page:
num_page_items = len(buyers)
for i in range(num_page_items):
    print(buyers[i].text + " : " + prices[i].text)

# Clean up (close browser once completed task).
wd.close()
```

```
# Define number of pages on the website
MAX_PAGE_NUM = 5
MAX_PAGE_DIG = 3
```

```
with open('results.csv', 'w') as f:
    f.write("Buyer, Price \n")

for i in range(1, MAX_PAGE_NUM + 1):
    page_num = (MAX_PAGE_DIG - len(str(i))) * "0" + str(i)
    url = "http://econpy.pythonanywhere.com/ex/" + page_num + ".html"

    wd.get(url)

    buyers = wd.find_elements_by_xpath('//div[@title="buyer-name"]')
    prices = wd.find_elements_by_xpath('//span[@class="item-price"]')

    num_page_items = len(buyers)
    with open('results.csv', 'a') as f:
        for i in range(num_page_items):
            f.write(buyers[i].text + "," + prices[i].text + "\n")

wd.close()
```
